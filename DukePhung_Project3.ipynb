# Assign the query name to variable.  Then use the variable as a paramter to .read_sql().  Close the connection and display the first 5 rows of the result.

```python
query = """select c.lastname, c.firstname, t.name, al.title 
from artist ar
join album al on ar.artistid = al.artistid 
join track t on al.albumid = t.albumid 
join invoiceline il on t.trackid = il.trackid 
join invoice i on il.invoiceid = i.invoiceid 
join customer c on i.customerid = c.customerid 
order by c.lastname, c.firstname;"""

connection = sqlite3.connect('chinnook.sqlite')
result = pd.io.sql.read_sql(query, connection)
connection.close()
result.head()
```
